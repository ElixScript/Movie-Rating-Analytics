```python
sns.set_style('whitegrid')
list1 = list()
mylabel = list()
for gen in movies.Genre.cat.categories :
    list1.append(movies[movies.Genre == gen].BudgetMillions)
    mylabel.append(gen)
fig , ax = plt.subplots()
fig.set_size_inches(11.7,8.27) # a4 paper 
h = plt.hist(list1,bins = 15, stacked = True,rwidth = 1, label = mylabel)
plt.title('Movie Budget Distributions',fontsize = 25,fontname = 'Times new roman')
plt.ylabel('Number Of Movies',fontsize = 20,color = 'Red')
plt.xlabel('Budget',fontsize = 20,color = 'Green')
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.legend(prop={'size':15},frameon = True,fancybox = True,shadow = True,framealpha = 1)
plt.show()
```
